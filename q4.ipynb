```python
import numpy as np
from qiskit import QuantumCircuit,transpile
from qiskit_aer import Aer
```

```python
def generate_random_circuit(n_qubits,depth):
    circuit = QuantumCircuit(n_qubits,n_qubits)
    for _ in range(depth):
        for qubit in range(n_qubits):
            circuit.rx(np.random.uniform(0,2*np.pi),qubit)
            circuit.ry(np.random.uniform(0,2*np.pi),qubit)
            circuit.rz(np.random.uniform(0,2*np.pi),qubit)
        for qubit in range(n_qubits-1):
            circuit.cz(qubit,qubit+1)   
    return circuit
```

```python
def random_benchmarking(n_qubits,depth,n_seq,shots):
    backend = Aer.get_backend('statevector_simulator')
    results = []
    for depth in depths:
        success_counts = 0
        for _ in range(n_seq):
            circuit = generate_random_circuit(n_qubits,depth)
            inverse_circuit = circuit.inverse()

            circuit_res = backend.run(circuit,shots=shots).result()
            final_statevector = circuit_res.get_statevector()

            inverse_res = backend.run(inverse_circuit,shots=shots).result()
            inverse_statevector = inverse_res.get_statevector()

            fidelity = np.abs(np.dot(inverse_statevector,final_statevector.conj()))**2
            success_counts += fidelity

        success_rate = success_counts/(n_seq*shots)
        results.append(success_rate)
    return results
```

```python
n_qubits = 2
depths = [1,2,3,4,5]
n_seq = 100
shots = 1000
```

```python
results = random_benchmarking(n_qubits,depths,n_seq,shots)
```
